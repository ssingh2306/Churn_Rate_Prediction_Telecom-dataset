# Feature Engineering

```python
data.isna().sum()
```

```python
data['TotalCharges'] = data.TotalCharges.fillna(value=data.TotalCharges.median())
```

```python
data.TotalCharges.median()
```

```python
columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

for column in columns:
    data[column] = data[column].replace({'No internet service': 'No'})
    print(f"Count of 'No internet service' in {column}: {data[column].value_counts().get('No', 0)}")
```

```python
columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']
for column in columns:
    data[column] = data[column].replace({'No internet service':'No'})
```

```python
data.head()
```

```python
binary_columns = [column for column in data.columns if data[column].nunique() == 2]
len(binary_columns)
```

```python
columns = ['Partner', 'Dependents', 'PhoneService', 'OnlineSecurity', 'OnlineBackup',  
           'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
           'PaperlessBilling', 'Churn']
for col in columns:
    data[col] = data[col].map({'Yes':1, 'No':0})
    
data['gender'] = data.gender.map({'Male':1, 'Female':0})
```

```python
for col in binary_columns:
    print(data[col].unique())
```

```python
customerID = data.customerID
data.drop('customerID', axis=1, inplace=True)
```

```python
def tenure_lab(period) :
    
    if period <= 12 :
        return 1
    elif (period > 12) & (period <= 24 ):
        return 2
    elif (period > 24) & (period <= 36) :
        return 3
    elif (period > 36) & (period <= 48) :
        return 3
    elif (period > 48) & (period <= 60) :
        return 4
    elif period > 60 :
        return 5
    
data['tenure'] = data.tenure.apply(tenure_lab)
```

```python
data.tenure.value_counts()
```

```python
dummy_col = ['MultipleLines', 'InternetService', 'Contract', 'PaymentMethod', 'tenure']
data_1 = pd.get_dummies(data, columns=dummy_col, drop_first=True, dtype='uint8')
```

```python
data_1.head()
```

```python
plt.figure(figsize=(20, 18))
sns.heatmap(data_1.corr(), annot=False)
```

```python
data_1.drop('Churn', axis=1).corrwith(data.Churn).plot(kind='barh', figsize=(10, 7))
```

```python
print(data_1.shape)

# Remove duplicate Features
data_1 = data_1.T.drop_duplicates()
data_1 = data_1.T

# Remove Duplicate Rows
data_1.drop_duplicates(inplace=True)

print(data_1.shape)
```

# Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

X = data_1.drop('Churn', axis=1)
y = data_1.Churn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = MinMaxScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
X_std = scaler.transform(X)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

def evaluate(model, X_train, X_test, y_train, y_test):
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("TRAINIG RESULTS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")

    print("TESTING RESULTS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")
```

```python
lr_clf = LogisticRegression(solver='liblinear', penalty='l1')
lr_clf.fit(X_train_std, y_train)

evaluate(lr_clf, X_train_std, X_test_std, y_train, y_test)
```

```python
disp = ConfusionMatrixDisplay.from_estimator(
    lr_clf, X_test_std, y_test, 
    cmap='Blues', values_format='d', 
    display_labels=['Stay', 'Churn']
)
```

```python
disp = RocCurveDisplay.from_estimator(
    lr_clf, X_test_std, y_test
)
```

```python
stay = (y_train.value_counts()[0] / y_train.shape)[0]
churn = (y_train.value_counts()[1] / y_train.shape)[0]

print(f"Staying Rate: {stay * 100:.2f}%")
print(f"Churning Rate: {churn * 100 :.2f}%")
```

# Random Forest Classifier

```python
rand_forest = RandomForestClassifier(n_estimators=1000)
rand_forest.fit(X_train, y_train)

evaluate(rand_forest, X_train, X_test, y_train, y_test)
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier  # Import RandomForestClassifier from sklearn.ensemble
from sklearn.model_selection import GridSearchCV    # Import GridSearchCV for hyperparameter tuning
from sklearn.metrics import f1_score                # Import f1_score to evaluate classifier performance

# Define the RandomForestClassifier with default parameters
rf_clf = RandomForestClassifier(n_estimators=100)   # Initialize RandomForestClassifier with 100 estimators

# Define the parameter grid for GridSearchCV
param_grid = {                                      # Define a dictionary specifying hyperparameters and their values
    'n_estimators': [550, 600, 650],                # Number of trees in the forest
    'max_depth': [3, 5, 7, 10, 15, None],           # Maximum depth of the trees
    'min_samples_split': [2, 3, 10],                # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 3, 5, 7, 10],           # Minimum number of samples required to be at a leaf node
    'criterion': ["gini", "entropy"]                # Function to measure the quality of a split
}

# Initialize GridSearchCV
rf_grid_cv = GridSearchCV(estimator=rf_clf,          # Random forest classifier to use
                          param_grid=param_grid,    # Parameter grid for hyperparameter tuning
                          scoring="f1",             # Scoring metric to optimize (F1 score)
                          n_jobs=-1,                # Number of jobs to run in parallel (-1 uses all available cores)
                          verbose=1,                # Verbosity level (1 for messages)
                          cv=3)                     # Number of folds in cross-validation

# Perform grid search to find the best parameters
rf_grid_cv.fit(X_train, y_train)                    # Fit GridSearchCV to training data

# Retrieve the best parameters found by GridSearchCV
best_params = rf_grid_cv.best_params_               # Get the best parameters found by grid search
print(f"Best parameters: {best_params}")

# Create a new RandomForestClassifier with the best parameters
rf_clf = RandomForestClassifier(**best_params)      # Initialize a new RandomForestClassifier with the best parameters

# Fit the classifier to the training data
rf_clf.fit(X_train, y_train)                        # Train the classifier on the training data

# Evaluate the classifier
train_f1_score = f1_score(y_train, rf_clf.predict(X_train))   # Calculate F1 score on training data
test_f1_score = f1_score(y_test, rf_clf.predict(X_test))      # Calculate F1 score on testing data

# Print the F1 scores
print(f"Train F1 Score: {train_f1_score}")          # Print F1 score on training data
print(f"Test F1 Score: {test_f1_score}")            # Print F1 score on testing data

```

```python
disp = ConfusionMatrixDisplay.from_estimator(
    rf_clf, X_test, y_test, 
    cmap='Blues', values_format='d', 
    display_labels=['Stay', 'Churn']
)
```

# XGBoost Classifier

```python
xgb_clf = XGBClassifier()
xgb_clf.fit(X_train, y_train)
evaluate(xgb_clf, X_train, X_test, y_train, y_test)
```

```python
hyperparameter_grid = {
    'n_estimators': [75, 90, 100, 125], 
    'learning_rate' : [0.09, 0.1, 0.2], 
    'base_score' : [0.15],
    'colsample_bytree':[0.85, 0.95, 1],
    'colsample_bylevel':[0.85, 0.95, 1],
    'colsample_bynode':[0.85, 0.95, 1]
}

xgb_clf = XGBClassifier()
xgb_cv = GridSearchCV(
    estimator=xgb_clf, 
    param_grid=hyperparameter_grid, 
    cv=3, 
    scoring='f1', 
    n_jobs =-1, 
    verbose=1
)

xgb_cv.fit(X_train, y_train)

best_params = xgb_cv.best_params_
print(f"Best parameters: {best_params}")

xgb_clf = XGBClassifier(**best_params)
xgb_clf.fit(X_train, y_train)
evaluate(xgb_clf, X_train, X_test, y_train, y_test)
```

```python
disp = ConfusionMatrixDisplay.from_estimator(
    xgb_clf, X_test, y_test, 
    cmap='Blues', values_format='d', 
    display_labels=['Stay', 'Churn']
)
```

```python
def feature_imp(df, model):
    fi = pd.DataFrame()
    fi["feature"] = df.columns
    fi["importance"] = model.feature_importances_
    return fi.sort_values(by="importance", ascending=False)
```

```python
df = feature_imp(data_1.drop('Churn', axis=1), xgb_clf)
df.set_index('feature', inplace=True)
df.plot(kind='barh', figsize=(10, 6))
plt.title('Feature Importance according to XGBoost')
```
